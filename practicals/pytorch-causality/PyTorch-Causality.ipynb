```python
%matplotlib inline
from matplotlib import pyplot as plt

import torch
import torch.nn as nn
from torch.autograd import Variable

def plot_both(x, y, pred_x=None, pred_y=None, label=""):
    fig = plt.figure(figsize=(8, 4))
    fig.suptitle(label)
    
    plt.subplot(1, 2, 1)
    plt.plot(x.numpy(), y.numpy(), '.', label='Data X -> Y')
    if pred_x is not None:
        plt.plot(x.numpy(), pred_x.numpy(), '.', label='Pred X -> Y')
    plt.legend()
        
    plt.subplot(1, 2, 2)
    plt.plot(y.numpy(), x.numpy(), '.', label='Data Y -> X')
    if pred_y is not None:
        plt.plot(y.numpy(), pred_y.numpy(), '.', label='Pred Y -> X')
    plt.legend()
    
    plt.show()
    return None
```

Seems like nothing happened, but under-the-hood `PyTorch` creates a computational graph to calculate the gradient of the loss. Graph is constructed dynamically upon invocation and destroyed immediately after evaluation. Gradients get accumulated into the variables. So if you do backward then forward and then backward, the gradients will accumulate, so be careful.

## Optimizers

Let's define the optimizatino function:

```python
opt = torch.optim.Adam(net.parameters())
```

```python
for i in range(10000):
    # zero gradients
    opt.zero_grad()
    # forward pass
    loss = mse(net(Variable(x)), Variable(y))
    # backward pass
    loss.backward()
    # update the parameters
    opt.step()
    
    if (i % 1000) == 0:
        print(i, loss.data[0])
```

## Multilayer perceptron
This already exists, but it's good practice defining your own modules.

```python
class MLP(nn.Module):
    def __init__(self, n_inputs, n_hiddens, n_outputs):
        super(MLP, self).__init__()
        self.linear1 = nn.Linear(n_inputs, n_hiddens)
        self.linear2 = nn.Linear(n_hiddens, n_outputs)
        self.activation = nn.ReLU()
        
    def forward(self, x):
        return self.linear2(self.activation(self.linear1(x)))
```

Substituting our earlier example for MLP:

```python
net = MLP(2, 20, 1)
mse = nn.MSELoss()
opt = torch.optim.Adam(net.parameters())

for i in range(10000):
    # zero gradients
    opt.zero_grad()
    # forward pass
    loss = mse(net(Variable(x)), Variable(y))
    # backward pass
    loss.backward()
    # update the parameters
    opt.step()
    
    if (i % 1000) == 0:
        print(i, loss.data[0])
```

```python
net = nn.Linear(2, 1)
mse = nn.MSELoss()
opt = torch.optim.Adam(net.parameters())

for i in range(10000):
    x = torch.randn(32, 2)
    y = (3 * x).sum(1)

    # zero gradients
    opt.zero_grad()
    # forward pass
    loss = mse(net(Variable(x)), Variable(y)) 
    # backward pass
    loss.backward()
    # update the parameters
    opt.step()
    
    if (i % 1000) == 0:
        print(i, loss.data[0])
        
#

class MLP(torch.nn.Module):
    def __init__(self, sizes, final=None):
        super(MLP, self).__init__()
        layers = []
        for i in range(len(sizes) - 1):
            layers.append(torch.nn.Linear(sizes[i], sizes[i + 1]))
            if i != (len(sizes) - 2):
                layers.append(torch.nn.LeakyReLU())
        if final is not None:
            layers.append(final)
        self.net = torch.nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)
    
def train(net, mse, x, y, n_epochs=50):
    opt = torch.optim.Adam(net.parameters())
    
    for epoch in range(n_epochs):
        p = torch.randperm(x.size(0)).long()
        xp = x[p]
        yp = y[p]
        for i1 in range(0, x.size(0), 32):
            # sample data
            i2 = min(i1+32, x.size(0))
            xi, yi = xp[i1:i2], yp[i1:i2]
            # reset gradients
            opt.zero_grad()
            # forward pass
            loss = mse(net(Variable(xi)), Variable(yi))
            # backward pass
            loss.backward()
            # parameter update
            opt.step() 
    return net
```

# Causality

## Gaussian Case

```python
def example_gaussian(n = 1024):
    x = torch.randn(n, 1)
    y = x + torch.randn(n, 1)
    return x, y
```

```python
x, y = example_gaussian()
plot_both(x, y)
```

How to identify causal relationship here? The visual depiction shows there is no representation for us to determine that `y` was determined from `x`, i.e., the causal relationship is not identifiable.

## Uniform Case

```python
def example_uniform(n = 1024):
    x = torch.rand(n, 1)
    y = x + torch.rand(n, 1)
    return x, y

x, y = example_uniform()
plot_both(x, y)
```

$X \sim P_X$

$N \sim P_N$

$Y \leftarrow f(X) + N$

$X \not\perp N$

```python
net_x = train(MLP([1, 20, 1]), nn.MSELoss(), x, y)
net_y = train(MLP([1, 20, 1]), nn.MSELoss(), y, x)
```

```python
x, y = example_uniform()
plot_both(x, y,
         pred_x = net_x(Variable(x)).data,
         pred_y = net_y(Variable(y)).data)
```

In the first case, the residuals are constant with the predictions, in the second case, they are varying with the predictions.

## Computing Residuals

```python
res_x = y - net_x(Variable(x)).data
res_y = x - net_y(Variable(y)).data
```

```python
plt.plot(x.numpy(), res_x.numpy(), '.')
plt.plot(y.numpy(), res_y.numpy(), '.')
```

```python
plt.plot(x.numpy(), res_x.numpy(), '.')
plt.show()
plt.plot(y.numpy(), res_y.numpy(), '.')
plt.show()
```

## Measuring Statistical Dependence
Correlation will likely be the same for both charts, so that correlation coefficient will not be useful.
Recall:
Two random variables `x` and `y` are independent if:

$$
p(x,y) = p(x)p(y)
$$

We can see the marginal visualizations above, how to get the joint distribution? Permute the dimensions.

```python
plt.plot(y.numpy(), res_y.numpy(), '.')
p = torch.randperm(y.size(0)).long() # can only index tensors with long format
plt.plot(y.numpy(), res_y[p].numpy(), '.')
plt.show()
```

How do we compare between the two distributions? We **could** do a statistical test. But let's do a binary classifier for the two point classes. If our binary classifier does better than chance, then the distributions must be different.

```python
def dependence(x, y):
    # blue dataset p(x,y):
    d0 = torch.cat((x, y), 1) # concatenate a label
    l0 = torch.zeros(x.size(0), 1) # vector of zeros of same size as x
    
    # orange dataset p(x)*p(y):
    p = torch.randperm(x.size(0)).long()
    d1 = torch.cat((x, y[p]), 1)
    l1 = torch.ones((x.size(0), 1))
    
    d = torch.cat((d0, d1), 0)
    l = torch.cat((l0, l1), 0)
    
    net = train(MLP([2, 128, 1], nn.Sigmoid()), nn.BCELoss(), d, l)
    
    ## should train and test on separate sets
    
    return ((net(Variable(d)).data > 0.5).float() == l).float().mean() # for threshold, cast to float, compare to l, and average
    
```

```python
dependence(x, res_x), dependence(y, res_y)
```

## Conditional Independence

Suppose $X\leftarrow Z\rightarrow Y$

Then you want to to check if $X\not\perp Y \vert Z$. Need to find $X\perp Y\vert Z$.

```python
z = torch.randn(1000, 1)
x = (1.0 + z).exp() + torch.randn(1000, 1)
y = z.pow(2) + torch.rand(1000, 1) * 0.5
plt.plot(x.numpy(), y.numpy(), '.')
```

```python
net_x = train(MLP([1, 20, 1]), nn.MSELoss(), z, x)
net_y = train(MLP([1, 20, 1]), nn.MSELoss(), z, y)

res_x = x - net_x(Variable(z)).data
res_y = y - net_y(Variable(z)).data
```

Dependence between `x` and `y` is still very large:

```python
dependence(x, y)
```

However, dependence between residuals is much lower:

```python
dependence(res_x, res_y)
```

Still some outliers, but looks more like product of marginals than last chart:

```python
plt.plot(res_x.numpy(), res_y.numpy(), ".")
```

```python
x = torch.randn(1000, 1)
z = torch.randn(1000, 1)
net = MLP([2, 20, 1])
y = net(Variable(torch.cat((x, z), 1))).data
plt.plot(x.numpy(), y.numpy(), '.')
```

### NCC

```python
class NCC(nn.Module):
    def __init__(self):
        super(NCC, self).__init__()
        self.emb = MLP([2, 128, 128])
        self.clf = MLP([128, 128, 1], nn.Sigmoid())
        
    def forward(self, x):
        # the input x is a tensor of size "bs x n x 2"
        h = Variable(torch.zeros(x.size(0), 128))
        
        for i, xi in enumerate(x):
            # xi is a tensor of size "n x 2"
            h[i] = self.emb(xi).mean(0)
            
        return self.clf(h)
```

### Generate the Data

```python
X_f = torch.zeros(500, 1000, 2)
Y_f = torch.zeros(500, 1)
X_b = torch.zeros(500, 1000, 2)
Y_b = torch.ones(500, 1)

for i in range(500):
    x = torch.randn(1000, 1)
    z = torch.randn(1000, 1)
    net = MLP([2, 20, 1])
    y = net(Variable(torch.cat((x, z), 1))).data
    
    X_f[i] = torch.cat((x, y), 1)
    X_b[i] = torch.cat((y, x), 1)
    
X = torch.cat((X_f, X_b), 0)
Y = torch.cat((Y_f, Y_b), 0)
```

```python
ncc = NCC()
train(ncc, nn.BCELoss(), X, Y)
```
